```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import matplotlib.ticker as mticker
```

```python
#chargement du dataset

data_frame = pd.read_csv('/Users/bakarybengaly/Documents/AnalyseventesB2B/dataset techsales360.csv')
data_frame.head()
```

```python
#visualisation des types des données 
data_frame.dtypes
```

```python
#visualisation des types des données 
data_frame.dtypes
```

```python
#changement du type de la date d'objet en datetime
data_frame['date_transaction'] = pd.to_datetime(data_frame['date_transaction'])
data_frame.dtypes

```

```python
#visualisation des valeurs manquantes 
data_frame.isnull().sum()
```

```python
#remplacements des valeurs manquantes par "aucune"
data_frame['campagne_marketing'] = data_frame['campagne_marketing'].fillna('aucune')
data_frame.isnull().sum()
```

```python
# statistique descriptive 
data_frame.describe()
```

```python
#creation de colonnes (Marge, roi)
data_frame['marge_eur'] = data_frame['montant_net_eur'] - data_frame['cout_acquisition_eur']
data_frame['marge_eur']
```

```python
data_frame['annee'] = data_frame['date_transaction'].dt.year
data_frame['mois']  = data_frame['date_transaction'].dt.month
```

```python
ca_annee_mois = data_frame.groupby(['annee', 'mois'])['montant_net_eur'].sum().reset_index()

# Labels et couleurs
mois_labels = ['Jan','Fév','Mar','Avr','Mai','Jun','Jul','Aoû','Sep','Oct','Nov','Déc']
colors = ['#2196F3', '#4CAF50', '#FF9800']

# Graphique
fig, ax = plt.subplots(figsize=(12, 5))
ax.set_title("CA mensuel par année (2021–2023)", fontsize=14, fontweight='bold')

for i, annee in enumerate([2021, 2022, 2023]):
    data = ca_annee_mois[ca_annee_mois['annee'] == annee]
    ax.plot(data['mois'], data['montant_net_eur'],
            marker='o', label=str(annee),
            color=colors[i], linewidth=2.5)

ax.set_xticks(range(1, 13))
ax.set_xticklabels(mois_labels)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e6:.2f}M€'))
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_ylabel("Chiffre d'affaires")

plt.tight_layout()
plt.show()
```

```python
#chiffre d'affaire par region

ca_par_region = data_frame.groupby('region')['montant_net_eur'].sum().sort_values()

plt.figure(figsize=(9, 5))
plt.barh(ca_par_region.index, ca_par_region.values,
         color='steelblue', edgecolor='white')

for i, val in enumerate(ca_par_region.values):
    plt.text(val + 30000, i, f'{val/1e6:.1f}M€', va='center', fontsize=9)

plt.title("CA total par région", fontsize=14, fontweight='bold')
plt.xlabel("CA net (€)")
plt.tight_layout()
plt.show()
```

```python
# CA par catégorie produit
ca_par_categorie = data_frame.groupby('categorie_produit')['montant_net_eur'].sum().sort_values()

plt.figure(figsize=(9, 5))
plt.barh(ca_par_categorie.index, ca_par_categorie.values,
         color='mediumseagreen', edgecolor='white')

for i, val in enumerate(ca_par_categorie.values):
    plt.text(val + 30000, i, f'{val/1e6:.1f}M€', va='center', fontsize=9)

plt.title("CA total par catégorie produit", fontsize=14, fontweight='bold')
plt.xlabel("CA net (€)")
plt.tight_layout()
plt.show()

```

```python
# CA par canal d'acquisition
ca_par_canal = data_frame.groupby('canal_acquisition')['montant_net_eur'].sum().sort_values()

plt.figure(figsize=(9, 5))
plt.barh(ca_par_canal.index, ca_par_canal.values,
         color='steelblue', edgecolor='white')

for i, val in enumerate(ca_par_canal.values):
    plt.text(val + 30000, i, f'{val/1e6:.1f}M€', va='center', fontsize=9)

plt.title("CA total par canal d'acquisition", fontsize=14, fontweight='bold')
plt.xlabel("CA net (€)")
plt.tight_layout()
plt.show()
```

```python
# CA total par campagne marketing
ca_par_campagne = data_frame.groupby('campagne_marketing')['montant_net_eur'].sum().sort_values()

plt.figure(figsize=(9, 5))
plt.barh(ca_par_campagne.index, ca_par_campagne.values,
         color='steelblue', edgecolor='white')

for i, val in enumerate(ca_par_campagne.values):
    plt.text(val + 30000, i, f'{val/1e6:.1f}M€', va='center', fontsize=9)

plt.title("CA total par campagne marketing", fontsize=14, fontweight='bold')
plt.xlabel("CA net (€)")
plt.tight_layout()
plt.show()
```

```python
# Taux de conversion moyen par canal
conversion_canal = data_frame.groupby('canal_acquisition')['taux_conversion_pct'].mean().sort_values()

plt.figure(figsize=(9, 5))
plt.barh(conversion_canal.index, conversion_canal.values,
         color='mediumseagreen', edgecolor='white')

for i, val in enumerate(conversion_canal.values):
    plt.text(val + 0.1, i, f'{val:.1f}%', va='center', fontsize=9)

plt.title("Taux de conversion moyen par canal", fontsize=14, fontweight='bold')
plt.xlabel("Taux de conversion (%)")
plt.tight_layout()
plt.show()
```

```python
# CA total par produit
# CA total par produit
ca_par_produit = data_frame.groupby('produit')['montant_net_eur'].sum().sort_values()

plt.figure(figsize=(10, 6))
plt.barh(ca_par_produit.index, ca_par_produit.values,
         color='steelblue', edgecolor='black')

for i, val in enumerate(ca_par_produit.values):
    plt.text(val + 30000, i, f'{val/1e6:.1f}M€', va='center', fontsize=9)

plt.title("CA total par produit", fontsize=14, fontweight='bold')
plt.xlabel("CA net (€)")
plt.tight_layout()
plt.show()
```

```python
# Marge moyenne par produit
marge_par_produit = data_frame.groupby('produit')['marge_eur'].mean().sort_values()

plt.figure(figsize=(10, 6))
plt.barh(marge_par_produit.index, marge_par_produit.values,
         color='mediumseagreen', edgecolor='black')

for i, val in enumerate(marge_par_produit.values):
    plt.text(val + 50, i, f'{val/1e3:.1f}K€', va='center', fontsize=9)

plt.title("Marge moyenne par produit", fontsize=14, fontweight='bold')
plt.xlabel("Marge moyenne (€)")
plt.tight_layout()
plt.show()
```

```python
# Remise moyenne par segment client
remise_par_segment = data_frame.groupby('segment_client')['remise_pct'].mean().sort_values()

plt.figure(figsize=(9, 5))
plt.barh(remise_par_segment.index, remise_par_segment.values,
         color='coral', edgecolor='white')

for i, val in enumerate(remise_par_segment.values):
    plt.text(val + 0.1, i, f'{val:.1f}%', va='center', fontsize=9)

plt.title("Remise moyenne par segment client", fontsize=14, fontweight='bold')
plt.xlabel("Remise moyenne (%)")
plt.tight_layout()
plt.show()
```

```python
# Impact de la remise sur la marge
plt.figure(figsize=(9, 5))
plt.scatter(data_frame['remise_pct'], data_frame['marge_eur'],
            alpha=0.3, color='steelblue', s=10)

plt.title("Impact de la remise sur la marge", fontsize=14, fontweight='bold')
plt.xlabel("Taux de remise (%)")
plt.ylabel("Marge (€)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
